```python
import pandas as pd
student = pd.read_csv(r"C:\Users\Nikhil Pathak\OneDrive\Desktop\Python coding\Pandas learning\student_performance.csv")
student.head()
```

```python
# Add a column indicating if the student passed
student["Passed"] = student["exam_score"] > 33 
```

```python
total_students = student.shape[0]
print("Total number of students are:", total_students)
```

```python
students_study2 = student[student["study_hours_per_day"] > 2].shape[0]
print("Total number of students who study more than 2 hours:", students_study2)
student[student["study_hours_per_day"] > 2].sample(4)
```

```python
Passed_stu = student[student["Passed"]].shape[0]
print("Total number of students who passed:", Passed_stu)
student[student["Passed"]].sample(4)
```

```python
students_pass_2 = student[(student["Passed"]) & (student["study_hours_per_day"] > 2)].shape[0]
print("Total number of students who passed and study more than 2 hours:", students_pass_2)
student[(student["Passed"]) & (student["study_hours_per_day"] > 2)].sample(4)
```

```python
percent_passed_given_2 = (students_pass_2 / students_study2) * 100
print("Percentage of passed students among those who study > 2 hours:", percent_passed_given_2)
```

```python
p_passed = student["Passed"].mean()
p_given_study2 = (students_pass_2 / students_study2)
p_study_2 = (student["study_hours_per_day"] > 2).mean()
p_study_2h_given_passed = (p_given_study2 * p_study_2) / p_passed
print("P(Studied > 2h | Passed):", p_study_2h_given_passed)
```
