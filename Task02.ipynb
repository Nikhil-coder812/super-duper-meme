```python
numeric_col = students.select_dtypes(include=np.number).columns
students[numeric_col].hist(figsize=(12,8),bins = 20,edgecolor = "black")
plt.suptitle("Distribution of Numeric Columns")
plt.tight_layout()
plt.show()
```

```python
if 'Addicted_Score' in students.columns:
    counts = students['Addicted_Score'].value_counts()
    plt.figure(figsize=(6,4))
    plt.bar(counts.index, counts.values, color='skyblue', edgecolor='black')
    plt.title("Addiction Level Distribution")
    plt.xlabel("Addicted_Score")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
if 'Avg_Daily_Usage_Hours' in students.columns and 'Addicted_Score' in students.columns:
    grouped = students.groupby('Addicted_Score')['Avg_Daily_Usage_Hours'].mean()

    plt.bar(grouped.index, grouped.values, color='orange', edgecolor='black')
    plt.title("Average Time Spent vs Addicted Score")
    plt.xlabel("Addicted Score")
    plt.ylabel("Average Time Spent (hrs)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
students
```
